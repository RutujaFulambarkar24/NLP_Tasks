# 1. Conceptual Understanding

1. Difference between "Love" and "love" in NLP.
In NLP, "Love" and "love" are treated as different tokens if the text is case-sensitive. This can increase vocabulary size and create inconsistency in the data. Therefore, text is usually converted to lowercase during preprocessing to maintain uniformity.

2. What happens if stopwords are not removed?
If stopwords like "the", "is", and "and" are not removed, they add unnecessary noise to the dataset. This can reduce the efficiency of machine learning models by making it harder to focus on important and meaningful words.

3. Two real-world scenarios where removing stopwords can be harmful.
Sentiment Analysis: Removing words like "not" can completely change the meaning of a sentence. For example, "I am not happy" becomes "I am happy", which is incorrect.
Question Answering Systems: Words like "who", "what", and "how" are important to understand the intent of a question. Removing them can make the sentence meaningless.

4. Difference between stemming and lemmatization
Stemming is a technique that removes prefixes or suffixes from words to get their root form, but it may not always produce a valid word. Lemmatization, on the other hand, converts words into their base form using proper linguistic rules and vocabulary, making it more accurate than stemming.

# 2. Build Advanced Preprocessing Function

```python
import re 
from collections import Counter

def preprocess_text(text):
    
    # handle empty or invalid input
    if not isinstance(text, str) or text.strip() == "":
        return [], ""
    
    # convert to lowercase
    text = text.lower()
    
    # remove urls
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # remove emails
    text = re.sub(r'\S+@\S+', '', text)
    
    # remove numbers
    text = re.sub(r'\d+', '', text)
    
    # handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # remove special characters (including emojis)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # tokenize
    tokens = text.split()
    
    # remove small words except 'no' and 'not'
    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]
    
    # join back to sentence
    clean_sentence = " ".join(tokens)
    
    return tokens, clean_sentence
```

# 3. Stress Testing

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this",
]

results = []

for text in sentences:
    tokens, clean = preprocess_text(text)
    
    print("Original Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print("-"*50)
    
    results.append(tokens)
```

# 4. Token Analytics

```python
for i, tokens in enumerate(results):
    
    if len(tokens) == 0:
        print(f"Sentence {i+1}: No tokens")
        continue
    
    total = len(tokens)
    unique = len(set(tokens))
    avg_length = sum(len(t) for t in tokens) / total
    
    print(f"Sentence {i+1}")
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg_length, 2))
    print("-"*40)
```

1. Which sentence had the most noise?
The sentence "Win $$$ now!!! Limited offer!!!" had the most noise.

2. Which sentence retained the most meaningful tokens after cleaning?
The sentence "I am not happy with this" retained the most meaningful tokens.

# 5. Frequency Analysis

```python
all_tokens = []

for t in results:
    all_tokens.extend(t)

counter = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

# 6. Build Full Pipeline

```python
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, clean = preprocess_text(text)
        
        all_tokens.extend(tokens)
        clean_sentences.append(clean)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
output = full_pipeline(sentences)

print("All Tokens:", output["tokens"])
print("\nClean Sentences:", output["clean_sentences"])
```

# 7. Error Handling

```python
print(preprocess_text(""))
print(preprocess_text("😂😂😂"))
print(preprocess_text("123456"))
```

The preprocessing function is designed to handle edge cases such as empty strings, text containing only emojis, and numeric-only input. In such cases, the function returns empty tokens and an empty cleaned sentence, ensuring robustness and preventing runtime errors.
